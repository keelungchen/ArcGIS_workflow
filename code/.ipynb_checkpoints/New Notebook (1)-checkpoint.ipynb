```python
import arcpy

# 設定空間參考 (WGS 1984)
spatial_ref = arcpy.SpatialReference(4326)
# 建立線段座標 [赤道, 北回歸線, 南回歸線]
latitudes = [0, 23.5, -23.5]
lines = []

for lat in latitudes:
    pts = arcpy.Array([arcpy.Point(-180, lat), arcpy.Point(180, lat)])
    lines.append(arcpy.Polyline(pts, spatial_ref))

# 匯出成 Feature Class (請修改路徑)
arcpy.CopyFeatures_management(lines, r"D:\ArcGIS\Morpho_Spatial_Demography\All_map.gdb\World_Lines.shp")
```
