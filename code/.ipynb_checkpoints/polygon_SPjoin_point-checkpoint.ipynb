```python
# ---------------------------------------------
# Python script for ArcGIS Pro Notebook
# Calculate, for each detect polygon, how many grid cells overlap,
# find the most frequent CNet_label_filtered among those cells,
# and compute the percentage that this mode label represents.
# ---------------------------------------------

import arcpy
from collections import Counter


# 1. Define your inputs
detect_fc = r"k2b_2508_corals"   # the DetectObjects polygon feature class
grid_fc   = r"rr_k2b_2025_preann_filtered"               # the 5cm grid polygon feature class
label_field = "species"         # the grid field to mode on

# 2. Create feature layers for selection operations
arcpy.management.MakeFeatureLayer(detect_fc, "detect_lyr")
arcpy.management.MakeFeatureLayer(grid_fc,   "grid_lyr")

# 3. Add fields to the detect layer if they do not already exist
#    - Join_Count (Long): number of overlapping grid cells
#    - Mode_Label (Text): most frequent label among those cells
#    - Pct_Mode (Double): percentage that the mode label represents
existing_fields = [f.name for f in arcpy.ListFields("detect_lyr")]
fields_to_add = [
    ("Join_Count", "LONG", None),
    ("Mode_Label", "TEXT", 50),
    ("Pct_Mode",  "DOUBLE", None)
]
for name, ftype, length in fields_to_add:
    if name not in existing_fields:
        if ftype == "TEXT":
            arcpy.management.AddField("detect_lyr", name, ftype, field_length=length)
        else:
            arcpy.management.AddField("detect_lyr", name, ftype)

# 4. Loop through each detect polygon, compute stats, and update fields
#    We'll use an UpdateCursor on the detect layer, accessing the shape object directly.
fields = ["OID@", "SHAPE@", "Join_Count", "Mode_Label", "Pct_Mode"]
with arcpy.da.UpdateCursor("detect_lyr", fields) as ucursor:
    for oid, geom, _, _, _ in ucursor:
        # Select grid cells that intersect this detect polygon
        arcpy.management.SelectLayerByLocation(
            "grid_lyr", "INTERSECT", geom, selection_type="NEW_SELECTION"
        )
        # Read the label values of the selected grid cells
        labels = [row[0] for row in arcpy.da.SearchCursor("grid_lyr", [label_field])]
        total = len(labels)
        
        if total > 0:
            # Determine the most common label and its count
            counts = Counter(labels)
            mode_label, mode_count = counts.most_common(1)[0]
            pct = (mode_count / total) * 100
        else:
            # No overlapping grids
            mode_label, mode_count, pct = None, 0, 0.0
        
        # Write back to the detect layer
        ucursor.updateRow([oid, geom, total, mode_label, pct])

# 5. Clear selections
arcpy.management.SelectLayerByAttribute("detect_lyr", "CLEAR_SELECTION")
arcpy.management.SelectLayerByAttribute("grid_lyr",   "CLEAR_SELECTION")

print("Finished computing overlap counts, mode labels, and percentages.")
```
