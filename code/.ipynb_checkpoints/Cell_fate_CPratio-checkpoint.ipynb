```python
import arcpy
import os

# === 路徑 & 參數 ===
gdb_path = r"D:\ArcGIS\Morpho_Spatial_Demography\MSD_Taiwan.gdb"
layer_t1 = os.path.join(gdb_path, "OL_P1_grid_4cm")
layer_t2 = os.path.join(gdb_path, "OL_P1_grid_4cm_2209")

key_field    = "SeqID"
id_field     = "Colony_ID"
fate_col_field = "Colony_Fate"  # 新增：紀錄珊瑚個體命運的欄位
fu_group_field = "Fu_group"     # 新增：紀錄融合群組 ID 的欄位

type_fields  = ["up_type", "down_type", "left_type", "right_type"]
fate_field   = "Cell_fate"
ratio_field  = "Cell_CPratio"

# 只計算的代碼（排除 Self / Ncoral）
target_codes = {"br","cb","di","co","ta","ma","en","fr","fo","bu","su"}

# === 確保欄位存在且型別正確 ===
existing = {f.name: f for f in arcpy.ListFields(layer_t1)}
if fate_field not in existing:
    arcpy.AddField_management(layer_t1, fate_field, "TEXT", field_length=10)

if ratio_field in existing and existing[ratio_field].type.upper() != "DOUBLE":
    arcpy.DeleteField_management(layer_t1, ratio_field)
if ratio_field not in {f.name for f in arcpy.ListFields(layer_t1)}:
    arcpy.AddField_management(layer_t1, ratio_field, "DOUBLE")

# === 讀 T2 做查表 ===
print("正在讀取 T2 資料建立索引...")
t2_map = {}
with arcpy.da.SearchCursor(layer_t2, [key_field, id_field] + type_fields) as cur:
    for seq, cid, up, down, left, right in cur:
        t2_map[seq] = {"Colony_ID": cid, "up": up, "down": down, "left": left, "right": right}

# === 核心計算 ===
print("開始判定 Cell_fate (包含 Fusion 邏輯)...")
# 更新欄位加入 Colony_Fate 與 Fu_group
upd_fields = [key_field, id_field, fate_col_field, fu_group_field] + type_fields + [fate_field, ratio_field]

with arcpy.da.UpdateCursor(layer_t1, upd_fields) as ucur:
    for row in ucur:
        seq_id   = row[0]
        cid_t1   = row[1]
        c_fate_t1 = str(row[2]) if row[2] else "" # Colony_Fate
        fu_group  = str(row[3]) if row[3] else "" # Fu_group (e.g. "1+2+3")
        
        # 取得 T1 四周鄰居類型
        t1_vals = [row[4], row[5], row[6], row[7]]

        # ------- Cell_fate 判定 -------
        fate = "Neutral"
        
        if seq_id not in t2_map:
            fate = "Death"
        else:
            cid_t2 = t2_map[seq_id]["Colony_ID"]
            
            # --- Fusion 融合處理邏輯 ---
            is_id_match = False
            
            # 如果 Colony_Fate 為 "6" (Fu)，檢查 T2 ID 是否在融合群組內
            if c_fate_t1 == "6" and fu_group:
                # 解析 "1+2+3" 為 ['1', '2', '3']
                fu_list = [str(i).strip() for i in fu_group.split('+')]
                if str(cid_t2) in fu_list:
                    is_id_match = True
            
            # 如果不是融合或融合 ID 沒對上，則進行標準 ID 比對
            if not is_id_match:
                if str(cid_t1) == str(cid_t2):
                    is_id_match = True
            
            # --- 最終 Fate 決定 ---
            if not is_id_match:
                fate = "Death"
            else:
                # ID 匹配成功 (不論是相同還是融合)，計算 Win/Lose
                t2_vals = [t2_map[seq_id]["up"], t2_map[seq_id]["down"],
                           t2_map[seq_id]["left"], t2_map[seq_id]["right"]]

                # Self 變化
                self_change = t2_vals.count("Self") - t1_vals.count("Self")
                
                # 計算競爭代碼 (target_codes) 變化
                others_t1 = sum(1 for v in t1_vals if v in target_codes)
                others_t2 = sum(1 for v in t2_vals if v in target_codes)
                delta = self_change - (others_t2 - others_t1)

                if   delta > 0: fate = "Win"
                elif delta < 0: fate = "Lose"
                else:           fate = "Neutral"

        # ------- Cell_CPratio 計算 -------
        count_target = sum(1 for v in t1_vals if v in target_codes)
        cp_ratio = count_target / 4.0

        # 寫回 UpdateCursor
        row[-2] = fate
        row[-1] = round(cp_ratio, 6)
        ucur.updateRow(row)

print("✅ 已完成：Cell_fate 判定 (已整合融合 Fusion 補償邏輯)")
```

```python
import arcpy
import os
import csv

# === 路徑 & 參數 ===
gdb_path = r"D:\ArcGIS\Morpho_Spatial_Demography\MSD_Taiwan.gdb"
layer_t1 = os.path.join(gdb_path, "OL_P1_grid_4cm")
layer_t2 = os.path.join(gdb_path, "OL_P1_grid_4cm_2209")

output_folder = r"D:\ArcGIS\Morpho_Spatial_Demography\data"
output_csv_name = "OL_P1_grid_4cm_results.csv"
output_csv_path = os.path.join(output_folder, output_csv_name)

key_field      = "SeqID"
id_field       = "Colony_ID"
fate_col_field = "Colony_Fate"
form_field     = "Form"  # 確保與資料庫欄位名一致
fu_group_field = "Fu_group"
type_fields    = ["up_type", "down_type", "left_type", "right_type"]
fate_field     = "Cell_fate"
ratio_field    = "Cell_CPratio"
fate_bi_field  = "Cell_fate_bi"

target_codes = {"br","cb","di","co","ta","ma","en","fr","fo","bu","su"}

# 🆕 定義對照轉換表 (Description Mapping)
# 使用字串作為 Key 以增加相容性
fate_desc_map = {
    "1": "D", "2": "G", "3": "S", "4": "Fi", "5": "GorS", "6": "Fu", "7": "Other"
}
form_desc_map = {
    "1": "br", "2": "cb", "3": "di", "4": "co", "5": "ta", "6": "ma",
    "7": "en", "8": "fr", "9": "fo", "10": "bu", "11": "su", "12": "NAN"
}

# === 1. 確保欄位存在 ===
existing = {f.name: f for f in arcpy.ListFields(layer_t1)}
if fate_field not in existing:
    arcpy.AddField_management(layer_t1, fate_field, "TEXT", field_length=10)
if ratio_field in existing and existing[ratio_field].type.upper() != "DOUBLE":
    arcpy.DeleteField_management(layer_t1, ratio_field)
if ratio_field not in {f.name for f in arcpy.ListFields(layer_t1)}:
    arcpy.AddField_management(layer_t1, ratio_field, "DOUBLE")
if fate_bi_field not in existing:
    arcpy.AddField_management(layer_t1, fate_bi_field, "SHORT")

# === 2. 讀取 T2 做查表 ===
print("正在讀取 T2 資料建立索引...")
t2_map = {}
with arcpy.da.SearchCursor(layer_t2, [key_field, id_field] + type_fields) as cur:
    for seq, cid, up, down, left, right in cur:
        t2_map[seq] = {"Colony_ID": cid, "up": up, "down": down, "left": left, "right": right}

# === 3. 核心計算 ===
print("開始執行獨立判定邏輯 (Cell_fate & Cell_fate_bi)...")
upd_fields = [key_field, id_field, fate_col_field, fu_group_field] + type_fields + [fate_field, ratio_field, fate_bi_field]

with arcpy.da.UpdateCursor(layer_t1, upd_fields) as ucur:
    for row in ucur:
        seq_id, cid_t1, c_fate_t1, fu_group = row[0], row[1], str(row[2]) if row[2] else "", str(row[3]) if row[3] else ""
        t1_vals = [row[4], row[5], row[6], row[7]]

        fate_bi = 0
        fate_str = "Neutral"

        if seq_id in t2_map:
            cid_t2 = t2_map[seq_id]["Colony_ID"]
            if str(cid_t1) == str(cid_t2):
                fate_bi = 1
            elif c_fate_t1 == "6" and fu_group:
                fu_list = [str(i).strip() for i in fu_group.split('+')]
                if str(cid_t2) in fu_list:
                    fate_bi = 1
        else:
            fate_bi = 0

        if fate_bi == 0:
            fate_str = "Death"
        else:
            t2_info = t2_map[seq_id]
            t2_vals = [t2_info["up"], t2_info["down"], t2_info["left"], t2_info["right"]]
            self_change = t2_vals.count("Self") - t1_vals.count("Self")
            others_t1 = sum(1 for v in t1_vals if v in target_codes)
            others_t2 = sum(1 for v in t2_vals if v in target_codes)
            delta = self_change - (others_t2 - others_t1)
            if delta > 0: fate_str = "Win"
            elif delta < 0: fate_str = "Lose"
            else: fate_str = "Neutral"

        count_target = sum(1 for v in t1_vals if v in target_codes)
        cp_ratio = count_target / 4.0

        row[-3], row[-2], row[-1] = fate_str, round(cp_ratio, 6), fate_bi
        ucur.updateRow(row)

# === 🆕 4. 匯出結果成 CSV (含 Description 轉換) ===
print(f"正在匯出結果至 CSV (執行代碼轉換): {output_csv_path} ...")

# 取得所有欄位清單 (排除幾何等欄位)
all_fields = [f.name for f in arcpy.ListFields(layer_t1) if f.type not in ["Geometry", "Raster", "Blob"]]

# 找出目標欄位在清單中的位置
try:
    fate_idx = all_fields.index(fate_col_field)
except ValueError:
    fate_idx = -1

try:
    form_idx = all_fields.index(form_field)
except ValueError:
    form_idx = -1



try:
    with open(output_csv_path, 'w', newline='', encoding='utf-8-sig') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(all_fields) # 寫入標題
        
        with arcpy.da.SearchCursor(layer_t1, all_fields) as s_cur:
            for row in s_cur:
                # tuple 不可修改，轉成 list 處理
                data_list = list(row)
                
                # 執行 Colony_Fate 轉換
                if fate_idx != -1:
                    val = str(data_list[fate_idx]) if data_list[fate_idx] is not None else ""
                    data_list[fate_idx] = fate_desc_map.get(val, val) # 找不到則保留原值
                
                # 執行 Form 轉換
                if form_idx != -1:
                    val = str(data_list[form_idx]) if data_list[form_idx] is not None else ""
                    data_list[form_idx] = form_desc_map.get(val, val)
                
                writer.writerow(data_list)
                
    print(f"✅ CSV 匯出成功！已自動將 Code 轉換為 Description。")
except Exception as e:
    print(f"❌ CSV 匯出失敗: {e}")

print("✅ 全部程序完成！")
```
