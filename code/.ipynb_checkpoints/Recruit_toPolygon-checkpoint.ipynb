```python
import arcpy
import collections
import os

def main():
    # ================= 參數設定區 (請依需求修改) =================
    input_layer = "OL_P1_2209_cor_SimplifyPolyg"  # 你的原始 Polygon 圖層名稱
    out_point_fc = "Recruit_Centroids"    # 輸出的中心點圖層名稱
    out_square_fc = "Recruit_Squares"     # 輸出的正方形圖層名稱
    
    # 欄位名稱對應 (原始欄位名稱)
    f_recruit = "Recruit"
    f_id = "ID"
    f_form = "Form"
    f_area = "Shape_Area"
    
    # 正方形邊長列表 (單位: 公分)
    square_sizes_cm = [2, 4, 8, 16, 32, 64, 128]
    
    # 設定環境
    arcpy.env.overwriteOutput = True
    # 取得輸入圖層的空間參考
    sr = arcpy.Describe(input_layer).spatialReference
    
    # 檢查座標系統單位 (為了正確轉換公分)
    unit_factor = 1.0
    if sr.linearUnitName.lower() in ['meter', 'metre']:
        unit_factor = 0.01  # 1 cm = 0.01 m
    elif sr.linearUnitName.lower() in ['foot', 'feet']:
        unit_factor = 0.0328084 # 1 cm = 0.0328 ft
    else:
        arcpy.AddWarning("⚠️ 警告：座標系統非公尺或英尺，請確認單位換算。目前假設單位為公尺。")
        unit_factor = 0.01

    # ================= 1. 處理 Domain (代碼轉文字) =================
    # 同時使用 print 以便在視窗立即查看進度
    msg = "正在讀取屬性域 (Domain) 資訊..."
    arcpy.AddMessage(msg)
    print(msg)
    
    # 獲取 Form 欄位使用的 Domain 名稱
    fields = arcpy.ListFields(input_layer)
    form_domain_name = None
    for field in fields:
        if field.name == f_form:
            form_domain_name = field.domain
            break
            
    # 建立 代碼->文字 的對照表
    domain_lookup = {}
    if form_domain_name:
        domains = arcpy.da.ListDomains(arcpy.env.workspace)
        for domain in domains:
            if domain.name == form_domain_name:
                if domain.domainType == 'CodedValue':
                    domain_lookup = domain.codedValues
                break
    else:
        warn_msg = f"注意：欄位 {f_form} 沒有設定 Domain，將直接使用原始值。"
        arcpy.AddWarning(warn_msg)
        print(warn_msg)

    # ================= 2. 生成中心點並轉移屬性 =================
    msg = "正在篩選 Polygon 並生成中心點..."
    arcpy.AddMessage(msg)
    print(msg)
    
    # 建立輸出點圖層 (使用 memory 暫存)
    create_result = arcpy.CreateFeatureclass_management("memory", out_point_fc, "POINT", spatial_reference=sr)
    out_path, out_name = os.path.split(create_result.getOutput(0))
    
    # 新增欄位 (注意：Shape_Area 改名為 Orig_Area 以免衝突)
    arcpy.AddField_management(out_point_fc, f_id, "TEXT") # 假設 ID 是文字，若是數字請改成 LONG
    arcpy.AddField_management(out_point_fc, "Form_Desc", "TEXT")
    arcpy.AddField_management(out_point_fc, "Orig_Area", "DOUBLE")

    # 準備統計數據
    stats_counter = collections.Counter()
    total_points = 0
    points_data = [] # 暫存點資料以便後續生成正方形

    # 搜尋游標 (讀取原始資料)
    fields_to_read = [f_recruit, "SHAPE@", f_id, f_form, f_area]
    
    with arcpy.da.SearchCursor(input_layer, fields_to_read) as search_cursor:
        # 插入游標 (寫入新點)
        fields_to_insert = ["SHAPE@", f_id, "Form_Desc", "Orig_Area"]
        with arcpy.da.InsertCursor(out_point_fc, fields_to_insert) as insert_cursor:
            
            for row in search_cursor:
                recruit_val = row[0]
                
                # 篩選條件：Recruit 為 "Y" (這裡加入 .strip() 去除可能存在的空白)
                if recruit_val and str(recruit_val).strip() == "Y":
                    geom = row[1]
                    oid = row[2]
                    raw_form = row[3]
                    area_val = row[4]
                    
                    # 取得幾何中心
                    centroid = geom.centroid
                    
                    # Domain 轉換
                    form_text = domain_lookup.get(raw_form, str(raw_form))
                    
                    # 寫入點圖層
                    insert_cursor.insertRow([centroid, oid, form_text, area_val])
                    
                    # 記錄數據供後續使用及統計
                    points_data.append({
                        "geom": centroid, 
                        "id": oid, 
                        "form": form_text, 
                        "area": area_val
                    })
                    stats_counter[form_text] += 1
                    total_points += 1

    msg = f"已生成 {total_points} 個中心點。"
    arcpy.AddMessage(msg)
    print(msg)

    if total_points == 0:
        print("沒有符合條件 (Recruit='Y') 的圖徵，程式結束。")
        return

    # ================= 3. 生成多層級正方形 =================
    msg = "正在生成多層級正方形..."
    arcpy.AddMessage(msg)
    print(msg)
    
    # 建立輸出正方形圖層
    arcpy.CreateFeatureclass_management(arcpy.env.workspace, out_square_fc, "POLYGON", spatial_reference=sr)
    
    # 新增欄位 (包含繼承的欄位 + Size_Level)
    arcpy.AddField_management(out_square_fc, f_id, "TEXT")
    arcpy.AddField_management(out_square_fc, "Form_Desc", "TEXT")
    arcpy.AddField_management(out_square_fc, "Orig_Area", "DOUBLE")
    arcpy.AddField_management(out_square_fc, "Size_Level_cm", "SHORT") # 記錄它是幾公分的大小

    # 插入游標 (寫入正方形)
    sq_fields = ["SHAPE@", f_id, "Form_Desc", "Orig_Area", "Size_Level_cm"]
    
    with arcpy.da.InsertCursor(out_square_fc, sq_fields) as cursor:
        for pt_data in points_data:
            center_pt = pt_data["geom"]
            cx, cy = center_pt.X, center_pt.Y
            
            for size_cm in square_sizes_cm:
                # 計算正方形的一半邊長 (換算成地圖單位)
                half_side = (size_cm * unit_factor) / 2.0
                
                # 建構正方形四個角點座標
                p1 = arcpy.Point(cx - half_side, cy - half_side)
                p2 = arcpy.Point(cx - half_side, cy + half_side)
                p3 = arcpy.Point(cx + half_side, cy + half_side)
                p4 = arcpy.Point(cx + half_side, cy - half_side)
                
                # 建立 Polygon 物件
                square_geom = arcpy.Polygon(arcpy.Array([p1, p2, p3, p4, p1]), sr)
                
                # 寫入資料
                cursor.insertRow([
                    square_geom, 
                    pt_data["id"], 
                    pt_data["form"], 
                    pt_data["area"], 
                    size_cm
                ])

    # ================= 4. 最終回報 (修正部分) =================
    report = ["\n" + "="*30]
    report.append(f"處理完成報告：")
    report.append(f"共使用了 {total_points} 個點生成正方形序列。")
    report.append("-" * 40)
    report.append(f"{'Form 類別':<15} | {'數量':<8} | {'比例':<8}")
    report.append("-" * 40)
    
    for form_type, count in stats_counter.most_common():
        # 安全計算比例，防止除以零
        percentage = (count / total_points) * 100
        report.append(f"{str(form_type):<15} | {count:<8} | {percentage:.1f}%")
        
    report.append("="*30 + "\n")
    
    # 輸出到 ArcGIS Pro 的訊息視窗與 Python 介面
    for line in report:
        arcpy.AddMessage(line)  # 給 Geoprocessing History
        print(line)             # 給 Python Window / Notebook (這是您缺少的)

if __name__ == '__main__':
    main()
```
